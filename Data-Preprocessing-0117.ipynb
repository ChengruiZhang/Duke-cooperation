## Matrix generation

```python
# generate matrix for IP 1(S2),2,3
def Generate_matrix_IP123(data):
    # sort by time, chip, identification, param
    # input: DataFrame
    # output: DataFrame*3, transformed data matrix, unique time matrix and STD matrix for duplicate cases
    D17 = data['D17'].drop_duplicates().sort_values().tolist()
    Matrix = pd.DataFrame(columns=['D18', 'D7', 'D14', 'D13', 'D19', 'D27', 'D20', 'D21'] + D17)
    a14_19 = data[['D18', 'D7', 'D14', 'D13', 'D19', 'D27', 'D20', 'D21']].drop_duplicates()
    for i in ['D18', 'D7', 'D14', 'D13', 'D19', 'D27', 'D20', 'D21']:
        Matrix[i] = a14_19[i]
    Unique_time = data['D14'].unique()
    STD = copy.deepcopy(Matrix)
    return Matrix, Unique_time, STD
```

```python
# 对于同一个时间，同一个芯片下的同一个独立标识，根据该标识对应的全部测试项是否仅有一个，生成结果矩阵
def Preprocessing_IP123(data, name, loc='/home/hs5/results/'):
    # for IP 1(S2),2,3
    # input: DataFrame, save_name, save_loc
    Matrix, Unique_time, STD = Generate_matrix_IP123(data)
    count, M_count, matrix_list, STD_list, ALL_LEN = 0, 0, [], [], Matrix.shape[0]
    print(ALL_LEN)
    # for the same time
    for i in Unique_time:
        if M_count % 100 == 0:
            print(M_count/ALL_LEN)
#             break
        tmp = data[data['D14'] == i]
        # for the same chip
        D18 = tmp['D18'].unique()
        for j in D18:
            tmp2 = tmp[tmp['D18'] == j]
            # for the same identification
            D19 = tmp2['D19'].unique()
            for k in D19:
                a2 = Matrix.iloc[M_count]
                STD_a2 = STD.iloc[M_count]
                tmp3 = tmp2[tmp2['D19'] == k]
                D17 = tmp3['D17'].unique()
#                 count_tmp = tmp3.shape[0]
                for n in D17:
                    tmp4 = tmp3[tmp3['D17'] == n]
                    if tmp4.__len__() > 1:
                        tmp5 = data.iloc[count:count + tmp4.__len__()]['VALUE']
                        a2[data.iloc[count]['D17']] = tmp5.mean()
                        STD_a2[data.iloc[count]['D17']] = tmp5.std()
                        count += tmp4.__len__()
                    else:
                        a2[data.iloc[count]['D17']] = data.iloc[count]['VALUE']
                        count += 1
                del tmp3
                M_count += 1
                matrix_list.append(a2)
                STD_list.append(STD_a2)
            del tmp2
        del tmp
    Matrix = pd.concat(matrix_list, axis = 1)
    Matrix = Matrix.transpose()
    STD = pd.concat(STD_list, axis = 1)
    STD = STD.transpose()
    Matrix.to_csv(loc + name + '.csv')
    STD.to_csv(loc + 'STD_' + name + '.csv')
```

```python
# generate matrix for IP1S1
def Generate_matrix_IP1(data):
    # sort by time, chip, identification, param
    # input: DataFrame
    # output: DataFrame*3, transformed data matrix, unique time matrix and STD matrix for duplicate cases
    D17 = data['D17'].drop_duplicates().sort_values().tolist()
    Matrix = pd.DataFrame(columns=['D1', 'D3', 'D4', 'D5', 'D6', 'D8', 'D9', 'D10', 'D11', 'D13', 'D14']+D17)
    a14_19 = data[['D1', 'D3', 'D4', 'D5', 'D6', 'D8', 'D9', 'D10', 'D11', 'D13', 'D14']].drop_duplicates()
    for i in ['D1', 'D3', 'D4', 'D5', 'D6', 'D8', 'D9', 'D10', 'D11', 'D13', 'D14']:
        Matrix[i] = a14_19[i]
    Unique_time = data['D14'].unique()
    STD = copy.deepcopy(Matrix)
    return Matrix, Unique_time, STD
```

```python
# 对于同一个时间，同一个芯片下的同一个独立标识，根据该标识对应的全部测试项是否仅有一个，生成结果矩阵
def Preprocessing_IP1(data, name, loc='/home/hs5/results/'):
    # for IP 1(S1)
    # input DataFrame, save_name, save_loc
    Matrix, Unique_time, STD = Generate_matrix_IP1(data)
    count, M_count, matrix_list, STD_list, ALL_LEN = 0, 0, [], [], Matrix.shape[0]
    print(ALL_LEN)
    # for the same time
    for i in Unique_time:
        if M_count % 100 == 0:
            print(M_count/ALL_LEN)
#             break
        tmp = data[data['D14'] == i]
        # for the same chip
        D18 = tmp['D1'].unique()
        for j in D18:
            tmp2 = tmp[tmp['D1'] == j]
            # for the same identification
            D19 = tmp2['D10'].unique()
            for k in D19:
                a2 = Matrix.iloc[M_count]
                STD_a2 = STD.iloc[M_count]
                tmp3 = tmp2[tmp2['D10'] == k]
                D17 = tmp3['D17'].unique()
#                 count_tmp = tmp3.shape[0]
                for n in D17:
                    tmp4 = tmp3[tmp3['D17'] == n]
                    if tmp4.__len__() > 1:
                        tmp5 = data.iloc[count:count + tmp4.__len__()]['VALUE']
                        a2[data.iloc[count]['D17']] = tmp5.mean()
                        STD_a2[data.iloc[count]['D17']] = tmp5.std()
                        count += tmp4.__len__()
                    else:
                        a2[data.iloc[count]['D17']] = data.iloc[count]['VALUE']
                        count += 1
                del tmp3
                M_count += 1
                matrix_list.append(a2)
                STD_list.append(STD_a2)
            del tmp2
        del tmp
    Matrix = pd.concat(matrix_list, axis = 1)
    Matrix = Matrix.transpose()
    STD = pd.concat(STD_list, axis = 1)
    STD = STD.transpose()
    Matrix.to_csv(loc + name + '.csv')
    STD.to_csv(loc + 'STD_' + name + '.csv')
```
